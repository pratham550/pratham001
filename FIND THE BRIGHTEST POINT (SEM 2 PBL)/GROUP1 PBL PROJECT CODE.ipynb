```python
#Group 1
#Gargi Barve
#Rujul Modi
#Prathmesh Hajare
#Janhavi Nagare


import cv2    #here we are importing opencv library
import matplotlib.pyplot as plt   #here we are importing matplotlib library
import numpy as np   #here we are importing numpy library


#to read an image we use imread function of opencv
#by adding a 0 beside the name of the image we turn that image into a grayscale image
say = cv2.imread("sky.jpg")
img = cv2.imread('sky.jpg',0)


#to display the image in a graph format we use imshow function of matplotlib library
plt.imshow(img)
```

```python
#minMaxLoc function of opencv helps us to find the brightest and darkest points 
#in an image and it also give their coordinates on the image
cv2.minMaxLoc(img)
```

```python
#here we are copying the img in a new variable img2 so that we can create a circle around the brightest point
img2=np.copy(img)
```

```python
#here we are assigning the coordinate values of brightest point to the variables 
maxx = cv2.minMaxLoc(img2)[3][0]
maxy = cv2.minMaxLoc(img2)[3][1]
```

```python
#here we are giving the radius, center,color and thickness of the circle
center = (maxx,maxy)
radius = 200
color = (255,0,0)
t = 60
```

```python
#this function is used to draw a circle around a specific point on the image
image9= cv2.circle(img2,center,radius,color,t)
```

```python
#this function is again used to display the image in this case with the circle
plt.imshow(image9)
```
